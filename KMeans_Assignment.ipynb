## Use sns.FacetGrid to create a stacked histogram showing Out of State Tuition based on the Private column. 

```python

# sns.histplot(data=df, x='Outstate', hue='Private',bins=30)
# plt.show()

Facegrid= sns.FacetGrid(df, hue="Private", height=5,aspect=1.6)
Facegrid.map(sns.histplot, "Outstate")
Facegrid.add_legend()
plt.show()
```

## Use sns.FacetGrid to create a stacked histogram showing Grad.Rate based on the Private column. 

```python
Facegrid= sns.FacetGrid(df, hue="Private", height=5,aspect=1.6)
Facegrid.map(sns.histplot, "Grad.Rate")
Facegrid.add_legend()
plt.show()
```

## Notice how there seems to be a private school with a graduation rate of higher than 100%.What is the name of that school?

```python
higher_then_100 = df.loc[(df['Private'] == 'Yes') & (df['Grad.Rate'] > 100)]
higher_then_100
```

## Set that school's graduation rate to 100 then re-do the histogram visualization.

```python
# finding where is the private that are higher then  >100 then saving them to redo it by the index 
i = higher_then_100.index
# Set the graduation rate of the school to 100
df.loc[i, 'Grad.Rate'] = 100

```

```python
higher_then_100
```

```python
Facegrid= sns.FacetGrid(df, hue="Private", height=5,aspect=1.6)
Facegrid.map(sns.histplot, "Grad.Rate")
Facegrid.add_legend()
plt.show()
```

## K Means Model

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=2)
```

## Fit the model to all the data except for the Private label.

```python
kmeans.fit(df.drop('Private', axis=1))
```

## What are the cluster center vectors?

```python
kmeans.cluster_centers_
```

## Create a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.

```python
df['Cluster'] = np.where(df['Private'] == 'Yes', 1, 0) # if the condution return a True value it will give it 1 if False it will give it 0 
df[df['Private']=='No']
```

```python
df.head()
```

## Create a confusion matrix and classification report to see how well the Kmeans clustering worked without being given any labels.

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
confusion_matrix(df['Cluster'], kmeans.labels_)
```

```python
print(classification_report(df['Cluster'], kmeans.labels_))
```
